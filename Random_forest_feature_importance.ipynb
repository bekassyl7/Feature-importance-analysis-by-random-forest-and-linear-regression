```python
dataset = pd.read_csv("/gdrive/MyDrive/Colab Notebooks/MD_specimen_label_chart.xlsx - No El.csv")
dataset
x = dataset[['T', 't', 'TS', 'YS', 'HT']]
y = dataset['RA']
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=12)
rf = RandomForestRegressor(n_estimators=100)
rf.fit(X_train, y_train)
rf.feature_importances_
plt.barh(X_train.columns, rf.feature_importances_)
```

```python
sorted_idx = rf.feature_importances_.argsort()
sorted_idx = sorted_idx[::-1]

import matplotlib.pyplot as plt
plt.rcParams["font.family"] = "Times New Roman"
plt.xlabel('Features', fontsize = 12)
plt.ylabel('Feature Importance', fontsize = 12)
plt.bar(X_train.columns[sorted_idx], rf.feature_importances_[sorted_idx], color = ['blue', 'green', 'orange', 'red', 'yellow'])
plt.savefig("/content/Random_forest_MD_without_EL")
```

```python
y_pred_mlr= rf.predict(X_test)
from sklearn import metrics
meanAbErr = metrics.mean_absolute_error(y_test, y_pred_mlr)
meanSqErr = metrics.mean_squared_error(y_test, y_pred_mlr)
rootMeanSqErr = np.sqrt(metrics.mean_squared_error(y_test, y_pred_mlr))
print('Mean Absolute Error:', meanAbErr)
print('Mean Square Error:', meanSqErr)
print('Root Mean Square Error:', rootMeanSqErr)
```
